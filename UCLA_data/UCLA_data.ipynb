```python
# Assign the dataset path for definition
dataset_path = '/content/drive/MyDrive/UCLA_data/'

# Creating a Neo reader for Neuralynx (.Ncs)
reader = neo.io.NeuralynxIO(dirname=dataset_path)

# Read dataset as a Block (high-level structure for organizing data)
block = reader.read_block()

# Access the raw signal data
seg = block.segments[0]  # Take the first segment
signal = seg.analogsignals[0]  # Get the first signal
sampling_rate = signal.sampling_rate  # Get the sampling rate

print(f"Signal shape: {signal.shape}")
print(f"Sampling rate: {sampling_rate}")
```

```python
# Creating a bandpass filter
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a

# Applying bandpass filter
def bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    # Set padlen to a smaller value
    y = filtfilt(b, a, data, padlen=0) # Set padlen to 0
    return y

# Processing the data in batches/chunks (Since my Free Google Colab RAM is insufficient)
chunk_size = 1000000  # 1 million samples per batch/chunk
num_chunks = int(np.ceil(len(signal) / chunk_size))  # Number of chunks

filtered_signal = np.zeros_like(signal)  # Placeholder for filtered signal

# Import AnalogSignal from neo
from neo import AnalogSignal

for i in range(num_chunks):
    start = i * chunk_size
    end = min((i + 1) * chunk_size, len(signal))
    print(f'Processing chunk {i + 1}/{num_chunks}: samples {start} to {end}')

    # Apply filtering to the current chunk
    # Access the magnitude of the signal for filtering and
    # then create a new AnalogSignal with the filtered data and original units
    filtered_data = bandpass_filter(signal[start:end].magnitude, 300, 3000, sampling_rate.magnitude)
    filtered_signal[start:end] = AnalogSignal(filtered_data, units=signal.units, sampling_rate=sampling_rate)

# filtered_signal now contains the filtered data with the correct units (yay! I fixed the error)
```
